```python
img=keras.preprocessing.image.load_img("dog.jpg",target_size=(256,256,3))
img
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
x=keras.preprocessing.image.img_to_array(img)
```

```python
type(x)
```

```python
type(img)
```

```python
i=0
for j in datagen.flow(x.reshape(1,256,256,3),batch_size=1,save_to_dir="aug"):
    i=i+1
    if i==10:
        break
```
