```python
df=pd.read_csv("/kaggle/input/credit-card-customer-churn-prediction/Churn_Modelling.csv")
df.head()
```

```python
df=df.iloc[:,3:]
df.head()
```

```python
df.shape
```

```python
df=pd.get_dummies(df,columns=["Geography",'Gender'])
```

```python
y=df.Exited
```

```python
X=df.drop(['Exited'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=742)
```

```python
xtrain_sc=sc.fit_transform(xtrain)
```

```python
xtest_sc=sc.transform(xtest)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(8,activation='tanh',input_dim=13))
model.add(Dense(8,activation='tanh'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

```python
model.fit(xtrain_sc,ytrain,epochs=25,batch_size=10)
```

```python
pred=model.predict(xtest_sc)
pred
```

```python
pred=np.where(pred>0.5,1,0)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pred)
```
